## –ú–æ–¥–µ–ª—å –ø—Ä–æ–≥–Ω–æ–∑–∏—Ä–æ–≤–∞–Ω–∏—è –æ—Ç—Ç–æ–∫–∞ –∫–ª–∏–µ—Ç–æ–≤ <a id="model"></a>

–†–∞–∑–¥–µ–ª–∏–º –¥–∞–Ω–Ω—ã–µ –Ω–∞ –ø—Ä–∏–∑–Ω–∞–∫–∏ (–º–∞—Ç—Ä–∏—Ü–∞ –•) –∏ —Ü–µ–ª–µ–≤—É—é –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é (y)

```python
X = gym_churn.drop('Churn', axis =1)
y = gym_churn['Churn']
```

–†–∞–∑–¥–µ–ª–∏–º –º–æ–¥–µ–ª—å –Ω–∞ –æ–±—É—á–∞—é—â—É—é –∏ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω—É—é –≤—ã–±–æ—Ä–∫–∏

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

–û–±—É—á–∏–º  –º–æ–¥–µ–ª—å –Ω–∞ train-–≤—ã–±–æ—Ä–∫–µ –¥–≤—É–º—è —Å–ø–æ—Å–æ–±–∞–º–∏:
- –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–µ–π,
- —Å–ª—É—á–∞–π–Ω—ã–º –ª–µ—Å–æ–º.

<div class="alert alert-success">
<b>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞:</b>

    
–í —Ü–µ–ª–æ–º –≤—ã–±–æ—Ä–∫–∞ —Ä–∞–∑–¥–µ–ª–µ–Ω–∞ –≤–µ—Ä–Ω–æ. –ú–æ–∂–Ω–æ —Ç–∞–∫–∂–µ –ø–æ–∫–∞–∑–∞—Ç—å –∏—Ç–æ–≥–æ–≤–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –Ω–∞—à–∏—Ö –¥–∞–Ω–Ω—ã—Ö.
    
–Ø –±—ã —Ä–µ–∫–æ–º–µ–Ω–¥–æ–≤–∞–ª –±—ã –µ—â—ë –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –∞—Ç—Ä–∏–±—É—Ç `stratify` –¥–ª—è —Ü–µ–ª–µ–≤–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π. –û–Ω –¥–∞–µ—Ç –Ω–∞–º –±–µ–∑–æ–ø–∞—Å–Ω–æ—Å—Ç—å —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏—è 20% –∑–Ω–∞—á–µ–Ω–∏–π —Ä–∞–≤–Ω–æ–≥–æ –Ω—É–∂–Ω–æ–≥–æ –Ω–∞–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∫–ª–∞—Å—Å–æ–≤ –≤ —Ü–µ–ª–µ–≤–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π.
    
</div>

```python
lg_model = LogisticRegression()
lg_model.fit(X_train, y_train)
lg_predictions = lg_model.predict(X_test)
```

```python
rf_model = RandomForestClassifier()
rf_model.fit(X_train, y_train)
rf_predictions = rf_model.predict(X_test)
```

–û—Ü–µ–Ω–∏–º –º–µ—Ç—Ä–∏–∫–∏ accuracy, precision –∏ recall –¥–ª—è –æ–±–µ–∏—Ö –º–æ–¥–µ–ª–µ–π –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ

```python
def metrics(y_true, y_pred, title = '–ú–µ—Ç—Ä–∏–∫–∏ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏'):
    print(title)
    print('\tAccuracy: {:.2f}'.format(accuracy_score(y_true, y_pred)))
    print('\tPrecision: {:.2f}'.format(precision_score(y_true, y_pred)))
    print('\tRecall: {:.2f}'.format(recall_score(y_true, y_pred)))
```

<div class="alert alert-success">
<b>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞:</b>

    
–û—á–µ–Ω—å –∫–ª–∞—Å—Å–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è —Ä–∞—Å—á–µ—Ç–æ–≤ –Ω—É–∂–Ω—ã—Ö –Ω–∞–º –º–µ—Ç—Ä–∏–∫ üëç
    
</div>
